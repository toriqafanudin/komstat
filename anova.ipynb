```r
# PEMBAHASAN SOAL NOMER 1
# Atur setwd() sesuai lokasi directory masing-masing
namaFile <- "/kaggle/input/anova1/soal1anova.csv" # Ubah sesuai nama file masing-masing
df <- read.csv(namaFile)

# Melihat susunan data
df
```

```r
# Melakukan uji normalitas
A <- df[df$Metode == "A", ]
B <- df[df$Metode == "B", ]
C <- df[df$Metode == "C", ]

shapiro.test(A$Nilai)
shapiro.test(B$Nilai)
shapiro.test(C$Nilai)
```

```r
# Melakukan uji homogenitas
nilai <- df$Nilai
metode <- df$Metode
bartlett.test(x=nilai, g=metode)
```

```r
# Melakukan uji anova
# H1: Terdapat perbedaan efektifitas antar metode pembelajaran
hasil_anova <- aov(nilai ~ metode, data = df)
summary(hasil_anova)
```

```r
# Menentukan titik kritis atau F tabel
alpha <- 0.05
dk1 <- 2
dk2 <- 40
titik_kritis <- qf(1-alpha, dk1, dk2)
titik_kritis
```

Dapat dilihat bahwa nilai F-value > titik kritis, sehingga dapat disimpulkan bahwa ada perbedaan efektifitas antar metode.

```r
# Cara melakukan uji normalitas pada residu (Optional)
res <- residuals(hasil_anova)
shapiro.test(res)
```

```r
# Melihat letak perbedaan (sama seperti post-hoc saat menggunakan SPSS)
TukeyHSD(hasil_anova)
```

Dapat dilihat bahwa p-adj untuk metode C-A adalah 0.036, kurang dari alpha 0.05, sehingga dapat dikatakan bahwa terdapat perbedaan efektifitas antara metode C dan A. Sementara untuk metode B-A dan C-B tidak ada perbedaan efektifitas.
