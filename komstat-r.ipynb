**PERHITUNGAN STATISTIK DESKRIPTIF**<br>
* Mean
* Median
* Standar Deviasi
* Summary

```r
data <- c(82, 77, 69, 70, 80, 78, 65, 68, 73, 84, 78, 73, 88, 85, 65)
rata_rata <- mean(data)
print(data)
print(rata_rata)
```

```r
nilaiMedian <- median(data)
print(nilaiMedian)
```

```r
nilaiSD <- sd(data)
print(nilaiSD)
```

```r
maksimal <- max(data)
minimal <- min(data)

print(maksimal)
print(minimal)
```

```r
keseluruhan <- summary(data)
print(keseluruhan)
```

**UJI NORMALITAS KOLMOGOROV-SMIRNOV**

Jika belum memiliki package nortest, diharuskan menginstal terlebih dahulu dengan cara:<br>
**install.packages("nortest")**

```r
library(nortest)
lillie.test(data)
```

1. Rumusan Hipotesis<br>
$H_0$: Data berdistribusi normal<br>
$H_1$: Data tidak berdistribusi normal<br>
2. $\alpha$ = 0.05<br>
3. Kriteria pengujian<br>
Jika **p-value > $\alpha$** maka $H_0$ diterima, dengan kata lain data berdistribusi normal<br>
4. Kesimpulan<br>
Berdasarkan output didapat nilai **p-value = 0.8742** yang mengakibatkan **p-value > $\alpha$**<br>
sehingga dapat disimpulkan $H_0$ diterima, atau data berdistribusi normal.
