ANALISIS REGRESI SEDERHANA

Akan digunakan soal UTS sebagai pembahasan, terlebih dahulu ubah data ke format csv.

```r
df = read.csv("/kaggle/input/hargarumah/hargaRumah.csv")
df
```

```r
harga = df$Harga
luasTanah = df$LuasTanah
jarakKeMall = df$JarakKeMall
model = lm(harga ~ luasTanah, data = df)
summary(model)
```

```r
plot(luasTanah, harga, main="Hubungan Luas Tanah dan Harga", xlab="Luas Tanah", ylab="Harga")
abline(model, col="red")
```

```r
koefisien = coef(model)
paste("koefisien =", koefisien)
# Persamaan Regresi
# prediksi_harga = koefisien[1] + luas_baru * koefisien[2]
luas_baru = 130
prediksi_harga = koefisien[1] + luas_baru * koefisien[2]
paste("prediksi harga =", prediksi_harga)
```

```r
# Uji Asumsi klasik yang diperlukan
#1 Uji Heteroskedastisitas (Breusch-Pagan Test)
# install.packages("lmtest")
library(lmtest)
bptest(model)
```

Jika p-value > alpha, maka varians residual dapat dianggap konstan dan asumsi homoskedastisitas dapat dianggap memenuhi.

```r
#2 Menguji apakah residual mengikuti distribusi normal
shapiro.test(residuals(model))
```
